# Notebook for XRD analysis

```python
# Import libraries
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Processing functions
def normalize(x):
    return (x-min(x))/(max(x)-min(x))
```

```python
# Import data
xls = pd.ExcelFile('./XRD/CuOx_XRD.xlsx')
df_cuo = pd.read_excel(xls, 'CuO')
df_cuox = pd.read_excel(xls, 'CuOx')
df_cuo_calc = pd.read_excel(xls, 'CuO_calc')
df_cuox_calc = pd.read_excel(xls, 'Cu2O_calc')
```

```python
df_cuox.plot(x='Two_Theta', y='Intensity')
df_cuo.plot(x='Two_Theta', y='Intensity')
```

```python
df_cuo = pd.merge(df_cuo, df_cuo_calc, how='inner', on='Two_Theta')
```

```python
df_cuo.plot(x='Two_Theta')
```

```python
df_cuo['Intensity_x'] = normalize(df_cuo['Intensity_x'])
```

```python
df_cuo['Intensity_y'] = normalize(df_cuo['Intensity_y']) 
```

```python
df_cuo.plot(x='Two_Theta')
plt.ylabel('Normalized Intensity')
plt.xlabel('Two-Theta (deg)')
plt.legend(['CuO Powder', 'CuO Calculated'])
plt.savefig('CuO_XRD.png', dpi=300)
```

```python
df_cuox = pd.merge(df_cuox, df_cuox_calc, how='inner', on='Two_Theta')
df_cuox.plot(x='Two_Theta')
```

```python
df_cuox = pd.merge(df_cuox, df_cuo_calc, how='inner', on='Two_Theta')
df_cuox.plot(x='Two_Theta')
```

```python
df_cuox['Intensity'] = normalize(df_cuox['Intensity'])
df_cuox['Intensity_x'] = normalize(df_cuox['Intensity_x'])
df_cuox['Intensity_y'] = normalize(df_cuox['Intensity_y'])
```

```python
df_cuox.plot(x='Two_Theta', y=['Intensity_x', 'Intensity', 'Intensity_y'])
plt.ylabel('Normalized Intensity')
plt.xlabel('Two-Theta (deg)')
plt.legend(['CuO$_x$ ALD', 'CuO Calculated', 'Cu$_2$O Calculated'])
plt.savefig('CuOx_XRD.png', dpi=300)
```
