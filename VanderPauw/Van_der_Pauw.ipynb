```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.interpolate import griddata
import re
from matplotlib.ticker import MaxNLocator
```

```python
time = [10, 10, 20, 30, 30]
temp = [250, 350, 300, 250, 350]
r = [2.06E10, 5.97E12, 2.13E12, 5.33E11, 1.44E12] 
```

```python
xi = np.arange(min(temp), max(temp), 1)
yi = np.arange(min(time), max(time)+1, 1)
xi,yi = np.meshgrid(xi,yi)
zi = griddata((temp, time), r, (xi, yi), method='linear')
```

```python
plt.contourf(xi, yi, zi)
plt.colorbar()
#plt.xticks(np.arange(1, 4, step=1))
plt.xlabel('Temp')
#plt.yticks(np.arange(1, 4, step=1))
plt.ylabel('Time')
plt.title('Linear Model: Sheet Resistance')
plt.savefig('vdpw', dpi=300)
plt.show()
```

```python
xi = np.arange(min(temp), max(temp), 1)
yi = np.arange(min(time), max(time)+1, 1)
xi,yi = np.meshgrid(xi,yi)
zi = griddata((temp, time), r, (xi, yi), method='cubic')

plt.contourf(xi, yi, zi/8.5E7)
plt.colorbar(format='%.0e')
#plt.xticks(np.arange(1, 4, step=1))
plt.xlabel('Anneal Temp. (C)')
#plt.yticks(np.arange(1, 4, step=1))
plt.ylabel('Anneal Time (min)')
plt.title('Cubic Model: Sheet Resistiviy (Ohm*cm)')
plt.ticklabel_format(style='sci')
plt.savefig('vdpw_cubic', dpi=300)
plt.show()
```
