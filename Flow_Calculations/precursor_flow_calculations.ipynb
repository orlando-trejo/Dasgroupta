# Notebook to calculate precursor flow rates

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Define functions and parameters

```python
# Antoine Equation
def antoine_eqn(A, B, C, T):
    ## T in K
    ## resulting pressure in bar
    return np.exp(A - B/(T+C))
```

```python
# Dalton's law of partial pressures
def precursor_flow(Fc, Pt, Vp):
    ## Fc: carrier gas flow rate
    ## Pt: total pressure
    ## Vp: precursor partial pressure (T)
    return Fc*(Vp/(Pt-Vp))
```

```python
# Parameters for DEZ
A = 4.41445
B = 1571.638
C = -34.978
T = np.arange(273, 373, 10)
Fc = [1, 2, 5, 10, 20] # carrier gas flow rate in slm
Pt = 1.01325 # total pressure in bar
```

## Calcualte and plot precursor flow rate

```python
# Flow rate for precursor
for flow in Fc:
    flow_mp = precursor_flow(flow, Pt, antoine_eqn(A, B, C, T))
    plt.plot(T-273, flow_mp)
    plt.xlabel('T (C)')
    plt.ylabel('Precusor Flow Rate (slm)')
    legend_labels = [('Carrier {} slm'.format(Fc[i])) for i in range(0, len(Fc)) ]
    plt.legend(legend_labels)
plt.show()
```
